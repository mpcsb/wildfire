```python
import math

def lat_long_to_cartesian(lat, long):
    # Earth's radius in km
    R = 6371
    # Convert latitude and longitude to radians
    lat = math.radians(lat)
    long = math.radians(long)
    # Calculate cartesian coordinates
    x = R * math.cos(lat) * math.cos(long)
    y = R * math.cos(lat) * math.sin(long)
    z = R * math.sin(lat)
    return x, y, z

# Path: testing\terrain_coordinates.ipynb
```

```python
import matplotlib.pyplot as plt

def plot_square(lat_long_coords, cartesian_coords):
    # Unpack the coordinates
    lat_coords, long_coords = zip(*lat_long_coords)
    x_coords, y_coords, z_coords = zip(*cartesian_coords)
    
    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2)
    
    # Plot the square in latitude and longitude
    ax1.plot(long_coords, lat_coords)
    ax1.set_title('Square in Latitude and Longitude')
    
    # Plot the square in cartesian coordinates
    ax2.plot(x_coords, y_coords)
    ax2.set_title('Square in Cartesian Coordinates')
    
    plt.show()
```

```python
# Define the coordinates of the square in latitude and longitude
lat_long_coords = [(0, 0), (0, 0.1), ( 0.1,  0.1), ( 0.1, 0), (0, 0)]

# Convert the coordinates to cartesian coordinates
cartesian_coords = [lat_long_to_cartesian(lat, long) for lat, long in lat_long_coords]

# Plot the square
plot_square(lat_long_coords, cartesian_coords)
```
